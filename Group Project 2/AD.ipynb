# make use of built-in julia AD tools

define a rosenbrock function

```julia
function Rosenbrock(x)
    sum = 0
    for i in 1:(size(x)[1]-1)
        sum += 100 * (x[i+1] - x[i]^2)^2. + (1-x[i])^2.
    end
    return sum
end

li = [5,4]

Rosenbrock(li)
```

Plot the rosenbrock function to confirm its functionality

```julia
#import Pkg
#Pkg.add("PythonPlot")
using Plots;
pythonplot()
x = range(-2,2, length=100)
y = range(-2,2, length=100)
rosen(x, y) = Rosenbrock([x, y])
contour(x, y, rosen)
```
